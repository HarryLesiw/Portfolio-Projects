```python
# import libraries
import pandas as pd
import seaborn as sns
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline

```

```python
# read in data as a data frame
df = pd.read_csv(r"C:\Users\hlesi\OneDrive - Loughborough University\Documents\Datasets\movies.csv")
df #shows a quick overview of data
```

```python
#Let us get some quick info about the dataset
#There are 7668 rows and 15 columns, we need to see if there are any missing data etc
df.info()
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Let's just drop the null values
df = df.dropna()
df.info()
```

```python
# Now look at the data types for each column
df.dtypes
```

```python
#Change data types to integers
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.dtypes
```

```python
# We're going to change the released column to just the year 
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df
```

```python
df = df.drop(columns = 'year') # deleting the year as it gave a few wrong  
df = df.sort_values(by = 'gross', ascending = False) #sort by grossing w/ highest first
df
```

```python
# Drop any duplicates
df.drop_duplicates()
```

```python
# Budget high correlation with revenue
# Company high correlation with revenue
```

```python
#Scatter plot with budget vs revenue
plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# plot budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {'color': 'red'}, line_kws = {'color': 'blue'})
```

```python
#Look at correlation
df.corr(numeric_only = True)
```

```python
correlation_matrix = df.corr(numeric_only = True)

sns.heatmap(data = correlation_matrix, annot = True)

plt.title('Correlation Matrix for Movie Aspects')
plt.xlabel('Movie Aspects')
plt.ylabel('Movie Aspects')
plt.show()
```

```python
#let us assign a number to of the entries in the columns
df_enumerate = df
for col_name in df_enumerate.columns:
    if df_enumerate[col_name].dtype == 'object':
        df_enumerate[col_name] = df_enumerate[col_name].astype('category')
        df_enumerate[col_name] = df_enumerate[col_name].cat.codes

df_enumerate
```

```python
df
```

```python
correlation_matrix = df_enumerate.corr(numeric_only = True)

sns.heatmap(data = correlation_matrix, annot = True)

plt.title('Correlation Matrix for Movie Aspects')
plt.xlabel('Movie Aspects')
plt.ylabel('Movie Aspects')
plt.rcParams['figure.figsize'] = (30,10)
plt.show()
```

```python
df_enumerate.corr()
```

```python
correlation_matrix2 = df_enumerate.corr()
corr_pairs = correlation_matrix2.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
